# **Functions**

### **concat** - when two dataframes need to concat vertically

```python
regs = pd.concat([nov,dec])
regs
```

*but we can see the index is also retained*

*`ignore_index = True` will solve this*

```python
regs = pd.concat([nov,dec],ignore_index=True)
regs
```

**Now if we don't want to remove the index instead we want to create a difference between these two datasets what can we do? We can use multindex dataframe**

```python
pd.concat([nov,dec],keys = ['Nov','dec'])
```

**If we want to stack horizontally**

```python
pd.concat([nov,dec],axis = 1)
```

### **merge** - joining datasets on a basis of one common column

Joining can be done by four ways:


1.   Inner Join
2.   Outer Join / Full join
3.   Left Join
4.   Right Join

##Untitled.png


**Inner Join** - Suppose i have two table and i want to join this then i'll have to find **common items** between them and join them in terms of that

To write code for this:
`leftTable.merge(rightTable,how='inner',on = on the basis of which column')`

```python
students.merge(regs,how = 'inner',on = 'student_id')
```

**Left Join** - This will join not only **common items** but also the uncommon items between them

To write code for this:
`leftTable.merge(rightTable,how='outer',on = on the basis of which column')`

```python
courses.merge(regs,how = 'left',on ='course_id')
```

**Left data irrespective of Right data mean left data will be printed even if right data isn't there**

**Right Join** - Right data irrespective of Left data mean Right data will be printed even if Left data isn't there

```python
temp_df = pd.DataFrame(
    {
        'student_id' : [26,27,28],
        'name' : ['Arian','Mahin','Sikder'],
        'partner':[28,26,17]

    }

)
```

```python
students = pd.concat([students,temp_df],ignore_index= True)
```

```python
students.tail()
```

```python
students.merge(regs,how='right',on = 'student_id')
```

**Outer Join** - It shows every data of left table and right table

```python
students.merge(regs,how='outer',on='student_id').tail(10)
```

**2. Find month by month revenue**

```python
temp_df = pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
```
