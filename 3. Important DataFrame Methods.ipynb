## **value_counts** - It counts frequency

*It is applicable in both series and dataframe*

```python
#Series
a = pd.Series([1,1,2,3,4,11,4,4,1,3,2,11])
a.value_counts()
```

```python
#DataFrame
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

marks
```

**Now if we use value counts in this dataframe, this will give us how many rows have repeated and the freq of their repeat.**

```python
marks.value_counts()
```

**In real life value_count() in dataframe is not useful it's useful in series**

```python
#Let's upload some datasets first to use this method
ipl = pd.read_csv('ipl-matches.csv')
ipl.head(2)
```

```python
#Find which player has won most potm in finals and qualifiers
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
#Toss decision plot -> After winning toss does the team bat or bowl
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
#How many matches each team has played
ipl['Team1'].value_counts() + ipl['Team2'].value_counts()
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
